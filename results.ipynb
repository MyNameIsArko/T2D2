```python
from tqdm.notebook import tqdm
import matplotlib.pyplot as plt
import numpy as np
```

```python
def get_episode_rewards(game):
    models = ['dqn', 'r2d2', 't2d2']
    episode_return = {}
    for model in tqdm(models):
        try:
            with open(f'logs/{model}/{game}/episode_return.txt', 'r') as file:
                rewards, env_steps = [], []
                for line in file:
                    reward, env_step = line.strip().split()
                    rewards.append(float(reward))
                    env_steps.append(int(env_step))
                episode_return[model] = {}
                episode_return[model]['x'] = np.array(env_steps)
                episode_return[model]['y'] = np.array(rewards)
        except:
            continue
    return episode_return
```

```python
cart_pole_return = get_episode_rewards('CartPole-v1')

plt.title('Cartpole episode reward during training')
plt.xlabel('Minutes of training')
plt.ylabel('Sum reward in given episode')

sma_dqn_y = np.convolve(cart_pole_return['dqn']['y'], np.ones(50)/50, mode='valid')
sma_r2d2_y = np.convolve(cart_pole_return['r2d2']['y'], np.ones(50)/50, mode='valid')
sma_t2d2_y = np.convolve(cart_pole_return['t2d2']['y'], np.ones(50)/50, mode='valid')

minutes_dqn = np.linspace(0, 4, len(sma_dqn_y))
minutes_r2d2 = np.linspace(0, 22, len(sma_r2d2_y))
minutes_t2d2 = np.linspace(0, 114, len(sma_t2d2_y))

plt.plot(minutes_dqn, sma_dqn_y, label='dqn')
plt.plot(minutes_r2d2, sma_r2d2_y, label='r2d2')
plt.plot(minutes_t2d2, sma_t2d2_y, label='t2d2')
plt.legend()
plt.show()
```
